```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
import numpy as np

def make_divider(W,B):
    divider = np.zeros((100,100))
    for row in range(0,99):
        for column in range(0,99):
           divider[row][column] = np.tanh((column * W + B - row)/10.)
    return divider

divider = make_divider(0,30)
imgplot = plt.imshow(divider)
divider2 = make_divider(1,0)
```

```python
divider2 = make_divider(0.5,30)
imgplot = plt.imshow(divider2)
```

```python
sum_divider = (divider)+(divider2)
imgplot = plt.imshow(sum_divider)
```

```python
divider3 = make_divider(-0.5,100)
imgplot = plt.imshow(divider3)
```

```python
sum_divider = (divider)+(divider2)+(divider3)
imgplot = plt.imshow(sum_divider)
```
